```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('convs.csv')
```

```python
df.columns = ['name', 'orders_count', 'order2payment','order2trial','trial2payment']
```

```python
df.iloc[:,1:2].hist(bins=20)
```

```python
nrows=2
ncols=2
ntt=0
fig, axs = plt.subplots(nrows=nrows, ncols=ncols,figsize=(15,15))
for i in range(nrows):
    for j in range(ncols):
        ntt+=1
        col = df.columns[ntt]
        axs[i,j].hist(df[col], bins=10)
        axs[i,j].set_title('Распределение {}'.format(col))
        axs[i,j].set_xlabel(col)
        axs[i,j].set_ylabel('Количество таких компаний')
        


```
